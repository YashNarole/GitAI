```python
##Stemming
from nltk.stem import PorterStemmer
```

```python
stemming=PorterStemmer()
print(stemming.stem("congratulations"))
```

```python
from nltk.stem import RegexpStemmer
```

```python
regexp_stemming=RegexpStemmer('Congratulating')
print(regexp_stemming)
print(regexp_stemming.stem("bulking"))
```

```python
import snowballstemmer
stemmer = snowballstemmer.stemmer('english')
words =['girl', 'girlling']
stemmed = stemmer.stemWords(words)
print(stemmed)
```
