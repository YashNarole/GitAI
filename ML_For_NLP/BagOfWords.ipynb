```python
# =====================================================
# BAG OF WORDS IMPLEMENTATION - SMSSpamCollection
# Step-by-Step Complete Pipeline
# =====================================================

# STEP 1: Import Required Libraries
import pandas as pd
import re
import numpy as np
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import CountVectorizer

# Download NLTK data (run once)
print("📥 Downloading NLTK data...")
nltk.download('stopwords')
print("✅ All libraries imported successfully!")

# STEP 2: Load Dataset with Exact File Path
file_path = r"C:\Users\PARAM 790\AI Playground\GitAI\ML_For_NLP\SMSSpamCollection.csv"
print(f"📂 Loading dataset from: {file_path}")

messages = pd.read_csv(file_path,
                      encoding="latin-1",
                      sep='\t',  # ✅ FIXED: Single quotes, not backslash
                      header=None,
                      names=["label", "message"])

print(f"✅ Dataset loaded! Shape: {messages.shape}")
print("\nFirst 5 rows:")
print(messages.head())

# STEP 3: Initialize Preprocessing Tools
print("\n🔧 Initializing preprocessing tools...")
ps = PorterStemmer()
stop_words = set(stopwords.words('english'))

# STEP 4: Text Preprocessing Pipeline
print("\n🧹 Cleaning text messages (this takes ~30 seconds)...")
corpus = []

for i in range(len(messages)):
    review = str(messages.at[i, 'message'])  # ✅ FIXED: Use .at[]
    
    review = re.sub('[^a-zA-Z]', ' ', review)
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if word not in stop_words]
    review = ' '.join(review)
    corpus.append(review)
    
    if (i+1) % 1000 == 0:
        print(f"Processed {i+1}/{len(messages)} messages")

print(f"✅ Preprocessing complete! Corpus size: {len(corpus)}")

# ✅ FIXED: Print section
print("\nSample cleaned messages:")
for i in range(3):
    original_msg = messages.at[i, 'message']
    print(f"Original: {str(original_msg)[:50]}...")
    print(f"Cleaned:  {corpus[i][:50]}...\n")

# STEP 5: Create Bag of Words Model
print("\n🎒 Creating Bag of Words matrix...")
vectorizer = CountVectorizer(max_features=2500)
X = vectorizer.fit_transform(corpus).toarray()

print(f"✅ Bag of Words matrix shape: {X.shape}")
print(f"   → {X.shape[0]} messages × {X.shape[1]} unique words")

# STEP 6: Create Final DataFrame with Labels
print("\n🏷️  Adding labels to Bag of Words...")
feature_names = vectorizer.get_feature_names_out()
bow_df = pd.DataFrame(X, columns=feature_names)
bow_df['label'] = messages['label']

print("✅ Final Bag of Words DataFrame ready!")
print("\nFinal DataFrame info:")
print(bow_df.info())
print("\nFirst 3 rows:")


# STEP 7: Save Results
print("\n💾 Saving results...")
bow_df.to_csv("SMSSpamCollection_BagOfWords.csv", index=False)
print("✅ Saved to: SMSSpamCollection_BagOfWords.csv")

# STEP 8: Quick Statistics
print("\n📊 Quick Stats:")
print(f"• Total messages: {len(messages)}")
print(f"• Vocabulary size: {len(feature_names)}")
print(f"• Sparse matrix density: {(X > 0).mean()*100:.1f}%")
print(f"• Ham messages: {(messages['label']=='ham').sum()}")
print(f"• Spam messages: {(messages['label']=='spam').sum()}")

print("\n🎉 BAG OF WORDS PIPELINE COMPLETE!")
```
